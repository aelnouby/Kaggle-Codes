```python
#imports
import pandas as pd
import numpy as np

#Random Forsts
from sklearn.ensemble import RandomForestClassifier
```

```python
train_csv=pd.read_csv('train.csv',header=0)
test_csv=pd.read_csv('test.csv',header=0)

test_csv.head()

```

```python
X_train=train_csv.drop(['label'],axis=1)
Y_train=train_csv.label

X_test=test_csv
```

```python
randomForest=RandomForestClassifier(n_estimators=100)
```

```python
randomForest.fit(X_train,Y_train)
predictions=randomForest.predict(X_test)
```

```python
submission=pd.DataFrame({'label':predictions})
submission['imageId']=submission.index+1
submission.to_csv('digit.csv',index=False)
```
